```python
import csv
import pandas as pd
df = pd.read_csv('./pyktok_output.csv')
```

```python
def merge_category_info(row):
    if not pd.isna(row['suggested_words']) and not pd.isna(row['diversificationLabels']):
        return f"{row['suggested_words']}, {row['diversificationLabels']}"
    elif not pd.isna(row['suggested_words']):
        return row['suggested_words']
    elif not pd.isna(row['diversificationLabels']):
        return row['diversificationLabels']
    else:
        return None

df['category_info'] = df.apply(merge_category_info,axis=1)
```

```python
df_dropped = df.dropna(subset=['category_info'])
```

```python
def determine_category(info):
    if 'sports' in info.lower() or 'fitness' in info.lower() or 'outdoor' in info.lower():
        return 'sports'
    elif 'beauty' in info.lower() or 'style' in info.lower():
        return 'beauty'
    elif 'society' in info.lower() or 'news' in info.lower() or 'social issues' in info.lower():
        return 'society'
    else:
        return 'Other' 
df_dropped['category'] = df_dropped['category_info'].apply(determine_category)
```

```python
import json

with open('./filtered_data.json') as f:
    dct = json.load(f)
dct[0]
```

```python
#rebuild dataset with the categories
rows = []
for entry in dct:
    distributed_link = entry['distributed_link']
    for rec_type, link in entry['rec_links'].items():
        # Extract the category from the rec_type if needed, e.g., 'beauty', 'sports', etc.
        category = rec_type.split('_')[0] 
        dist_id = distributed_link.split('/')[-1].split('?')[0]
        rec_id = link.split('/')[-1].split('?')[0]
        distr_category = 'Other'
        rec_category = 'Other'
        rec_type = rec_type.replace("control_society_rec", 'control_rec')
        try:
            distr_category = link_to_category_dict[int(dist_id)]
            rec_category = link_to_category_dict[int(rec_id)]
        except:
            pass
        rows.append({
            'distributed_link_id': dist_id,
            'distributed_link_category': distr_category,
            'persona_category': category,
            'rec_category': rec_category
        })

# Create DataFrame
df_temp = pd.DataFrame(rows)

# Example to print the DataFrame
#df[df['rec_category'].isna()] #95 bc filtered out x suggested/category info

df_cleaned = df_temp.groupby(['distributed_link_id', 'persona_category','distributed_link_category'])['rec_category'].agg(list).reset_index()
df_cleaned
```

```python
import numpy as np

persona = {'control':{'beauty': 0.20, 'sports': 0.10, 'society': 0.05, 'Other': 0.65},
    'sports': {'beauty': 1e-10, 'sports': 1, 'society': 1e-10, 'Other': 1e-10},
    'beauty': {'beauty': 1, 'sports': 1e-10, 'society': 1e-10, 'Other': 1e-10},
    'society': {'beauty': 1e-10, 'sports': 1e-10, 'society': 1, 'Other': 1e-10},
    'Other':{'beauty': 0.20, 'sports': 0.10, 'society': 0.05, 'Other': 0.65}
}

def compute_kl_divergence(recommended, persona):
    # Convert list of recommendations into a distribution
    rec_count = pd.Series(recommended).value_counts(normalize=True)
    rec_dist = rec_count.to_dict()
    
    # Ensure all categories are present in the distribution
    for category in persona.keys():
        if category not in rec_dist:
            rec_dist[category] = 0
    
    # Calculate KL divergence
    kl_div = 0
    for category, prob in persona.items():
        if rec_dist[category] == 0:
            rec_dist[category] = 1e-5  # Avoid log(0)
        kl_div += prob * np.log(prob / rec_dist[category])
    
    return kl_div

# Apply the KL divergence function to each row in the DataFrame
df_cleaned['KL_Divergence_persona'] = df_cleaned.apply(lambda row: compute_kl_divergence(row['rec_category'], persona[row['persona_category']]), axis=1)
df_cleaned['KL_Divergence_distr_link'] = df_cleaned.apply(lambda row: compute_kl_divergence(row['rec_category'], persona[row['distributed_link_category']]), axis=1)
```
