```python
import pandas as pd
import plotly.graph_objects as go
```

```python
df = pd.read_csv("cosinesim_recs_vs_original.csv")

# Convert 'current_time' column to datetime with varying formats
df['current_time'] = pd.to_datetime(df['current_time'], errors='coerce')

# Drop rows with invalid datetime values
df.dropna(subset=['current_time'], inplace=True)

# Define the personas
personas = df['persona'].unique()
```

```python
# Create a subplot for each persona
for persona in personas:
    # Filter df for each persona
    persona_df = df[df['persona'] == persona]
    
    # Sort df by 'current_time' 
    persona_df.sort_values(by='current_time', inplace=True)
    
    # Create a new figure for each persona
    fig = go.Figure()
    
    # Plot the time series for avg_cosim_allrecs
    fig.add_trace(go.Scatter(x=persona_df['current_time'], y=persona_df['avg_cosim_allrecs'],
                             mode='lines',
                             name=persona))
    # Update layout
    fig.update_layout(title=f'Average Cosine Similarity Over Time for {persona} Persona',
                      xaxis_title='Time',
                      yaxis_title='Average Cosine Similarity')


    fig.show()
```
