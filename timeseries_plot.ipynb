```python
import pandas as pd
import plotly.graph_objects as go
```

```python
data = pd.read_csv("cosinesim_recs_vs_original.csv")

# Convert 'current_time' column to datetime
data['current_time'] = pd.to_datetime(data['current_time'], format='%Y-%m-%d %H:%M:%S', errors='coerce')

# Check for any rows with NaT values
na_indices = data['current_time'].isna()
if na_indices.any():
    print("Rows with unparseable timestamps:")
    print(data[na_indices]['current_time'])

# Get unique personas
personas = data['persona'].unique()


```

```python

for persona in personas:
    persona_data = data[data['persona'] == persona]
    
    plt.figure(figsize=(10, 6))
    plt.plot(persona_data['current_time'], persona_data['avg_cosim_allrecs'], linestyle='-', color='black', label='Average')

    plt.title(f'Average Cosine Similarity Over Time for {persona} Persona')
    plt.xlabel('Time')
    plt.ylabel('Cosine Similarity')
    plt.legend()
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    
 
    plt.show()
```
